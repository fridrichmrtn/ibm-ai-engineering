### Step-by-Step Guide: 

#### Step 1: Setup the Environment 

Before we start, make sure to import the required libraries: TensorFlow and Keras. Keras is included within TensorFlow as `tensorflow.keras`. 


````python
#!pip install tensorflow==2.16.2 matplotlib==3.9.1
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.preprocessing.image import ImageDataGenerator
````

##### Explanation:
- `tensorflow` is the main library for machine learning in Python.
- `Sequential` is used to create a model with a linear stack of layers.
- `Dense` and `Flatten` are types of layers that we will use in our model.


#### Step 2: Load Pre-trained Model 


````python
# Load the VGG16 model pre-trained on ImageNet
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze the base model layers
for layer in base_model.layers:
    layer.trainable = False
````

#### Step 3: Create and Compile the Model 


````python
# Create a new model and add the base model and new layers
model = Sequential([
    base_model,
    Flatten(),
    Dense(256, activation='relu'),
    Dense(1, activation='sigmoid')  # Change to the number of classes you have
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
````

### **Create Placeholder Images**


````python
import os
from PIL import Image
import numpy as np

# Create directories if they don't exist
os.makedirs('sample_data/class_a', exist_ok=True)
os.makedirs('sample_data/class_b', exist_ok=True)

# Create 10 sample images for each class
for i in range(50):
    # Create a blank white image for class_a
    img = Image.fromarray(np.ones((224, 224, 3), dtype=np.uint8) * 255)
    img.save(f'sample_data/class_a/img_{i}.jpg')

    # Create a blank black image for class_b
    img = Image.fromarray(np.zeros((224, 224, 3), dtype=np.uint8))
    img.save(f'sample_data/class_b/img_{i}.jpg')

print("Sample images created in 'sample_data/'")
````

#### Step 4: Train the Model 


````python
# Load and preprocess the dataset
train_datagen = ImageDataGenerator(rescale=1./255)
train_generator = train_datagen.flow_from_directory(
    'sample_data',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

# Verify if the generator has loaded images correctly
print(f"Found {train_generator.samples} images belonging to {train_generator.num_classes} classes.")

# Train the model
if train_generator.samples > 0:
    model.fit(train_generator, epochs=10)
````

#### Step 5: Fine-Tune the Model 


````python
# Unfreeze the top layers of the base model 

for layer in base_model.layers[-4:]:
    layer.trainable = True 

# Compile the model again 
model.compile(optimizer='adam', loss='binary_crossentropy',
              metrics=['accuracy']) 

# Train the model again 
model.fit(train_generator, epochs=10) 
````

#### Exercise 1: Visualize Training and Validation Loss

**Objective:** Plot the training and validation loss to observe the learning process of the model.

**Instructions:**
1. Modify the training code to include validation data.
2. Plot the training and validation loss for each epoch.


````python
# img data
import glob
import pandas as pd

df = pd.DataFrame(glob.glob("sample_data/*/*"), columns=["path"])
df["label"] = df["path"].apply(lambda x: x.split("/")[1])
df["rnd"] = np.random.rand(df.shape[0])
df["set"] = df["rnd"].apply(lambda x: "train" if x < 0.6 else
                            ("validation" if x < 0.8 else "test"))

# data generators
generators = {"train": None, "validation": None, "test": None }
for set_name in df["set"].unique():
    set_df = df[df["set"] == set_name]
    datagen = ImageDataGenerator(rescale=1./255) # validation_split = 0.2
    generator = datagen.flow_from_dataframe(set_df, x_col="path", y_col="label",
                                          target_size=(224, 224), batch_size=32,
                                          class_mode="binary")
    generators[set_name] = generator
````

````python
# Load the VGG16 model pre-trained on ImageNet
base_model = VGG16(weights='imagenet', include_top=False,
                   input_shape=(224, 224, 3))

# Freeze the base model layers
for layer in base_model.layers:
    layer.trainable = False

# Create a new model and add the base model and new layers
model = Sequential([
    base_model,
    Flatten(),
    Dense(256, activation='relu'),
    Dense(1, activation='sigmoid')  # Change to the number of classes you have
])

model.compile(optimizer='adam', loss='binary_crossentropy',
              metrics=['accuracy'])

training_history = model.fit(generators["train"],
                             validation_data=generators["validation"],
                             epochs=10)
````

````python
import matplotlib.pyplot as plt

def plot_training_history(training_history, label):
    epochs = len(training_history.history["loss"])
    train_acc = training_history.history["accuracy"][-1]
    val_acc = training_history.history["val_accuracy"][-1]

    f,a = plt.subplots(1,1,figsize=(8, 5))
    pd.DataFrame(training_history.history).plot(y=["loss", "val_loss"], 
                                            ax=a)
    a.title.set_text('Train & validation loss over epochs, ' + label)
    a.set_xlabel("epochs")
    a.set_ylabel("loss")
    a.legend(["train", "validation"])
    a.set_xticks(range(0,epochs))
    a.set_xticklabels(range(1,epochs+1))
    f.tight_layout()
    print(f"train accuracy: {train_acc:.2f}, validation accuracy: {val_acc:.2f}")
    plt.show()

plot_training_history(training_history, "first model")
````

#### Exercise 2: Experiment with Different Optimizers

**Objective:** Experiment with different optimizers and observe their impact on model performance.

**Instructions:**
1. Change the optimizer from `adam` to `sgd` and `rmsprop`.
2. Retrain the model with each optimizer and compare the accuracy.


````python
optimizers = ["adam", "sgd", "rmsprop"]
training_histories = {}
for optimizer in optimizers:
    model = Sequential([
        base_model,
        Flatten(),
        Dense(256, activation='relu'),
        Dense(1, activation='sigmoid')])
        
    model.compile(optimizer=optimizer, loss='binary_crossentropy',
                  metrics=['accuracy'])
    
    training_histories[optimizer] = model.fit(generators["train"],
                                 validation_data=generators["validation"],
                                 epochs=10, verbose=0)
````

````python
for optimizer, training_history in training_histories.items():
    plot_training_history(training_history, optimizer)
````
